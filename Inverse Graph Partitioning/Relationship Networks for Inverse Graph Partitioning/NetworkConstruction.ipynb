## Network Construction

This file describes the methods to construct the relationship network.

single-layer netwok

```python
import pandas as pd
import networkx as nx
#choose detention(s)
detention = ['A','B','C','D']
#choose gender(s)
gender = ['M','F']
for deten in detention:
    for gen in gender:
        df = pd.read_csv(deten+'/'+deten+gen+'.csv',converters={'education':str,'cause':str})
        
        #record
        nodelist = list(df.id)
        df.index = df.id
        node_attrs_df = df.T
        node_attrs = node_attrs_df.to_dict()
        edgelist1 = []
        record_gp = df.groupby('record')
        for gp1 in record_gp:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        edgelist1.append((nlist[i],nlist[j]))           
        G1 = nx.Graph()
        G1.add_nodes_from(nodelist)
        G1.add_edges_from(edgelist1)
        nx.set_node_attributes(G1, node_attrs)
        nx.write_yaml(G1,deten+'/single-layer network/'+deten+gen+'_record.yaml')
        
        #cause
        edgelist2 = []
        record_gp = df.groupby('cause')
        for gp1 in record_gp:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        edgelist2.append((nlist[i],nlist[j]))           
        G2 = nx.Graph()
        G2.add_nodes_from(nodelist)
        G2.add_edges_from(edgelist2)
        nx.set_node_attributes(G2, node_attrs)
        nx.write_yaml(G2,deten+'/single-layer network/'+deten+gen+'_cause.yaml')
        
        #hometown
        edgelist3 = []
        record_gp = df.groupby('hometown')
        for gp1 in record_gp:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        edgelist3.append((nlist[i],nlist[j]))           
        G3 = nx.Graph()
        G3.add_nodes_from(nodelist)
        G3.add_edges_from(edgelist3)
        nx.set_node_attributes(G3, node_attrs)
        nx.write_yaml(G3,deten+'/single-layer network/'+deten+gen+'_hometown.yaml')

        ##network with all the relationships but unweighted
        edgelist = []
        record_gp1 = df.groupby('record')
        for gp1 in record_gp1:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        if (nlist[i],nlist[j]) not in edgelist:
                            edgelist.append((nlist[i],nlist[j]))
        record_gp2 = df.groupby('hometown')
        for gp1 in record_gp2:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        if (nlist[i],nlist[j]) not in edgelist:
                            edgelist.append((nlist[i],nlist[j]))
        record_gp3 = df.groupby('cause')
        for gp1 in record_gp3:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        if (nlist[i],nlist[j]) not in edgelist:
                            edgelist.append((nlist[i],nlist[j]))                
        G = nx.Graph()
        G.add_nodes_from(nodelist)
        G.add_edges_from(edgelist)
        nx.set_node_attributes(G, node_attrs)
        nx.write_yaml(G,deten+'/single-layer network/'+deten+gen+'_all_unweighted.yaml')
```

second-layer network

```python
import pandas as pd
import networkx as nx

#choose detention(s)
detention = ['A','B','C','D']
#choose gender(s)
gender = ['M','F']
for deten in detention:
    for gen in gender:
        df = pd.read_csv(deten+'/'+deten+gen+'.csv',converters={'education':str,'cause':str})
        nodelist = list(df.id)
        df.index = df.id
        node_attrs_df = df.T
        node_attrs = node_attrs_df.to_dict()

        # record+hometown
        edgelist = []
        record_gp = df.groupby('record')
        for gp1 in record_gp:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        edgelist.append((nlist[i],nlist[j],{'weight':3}))
        record_gp = df.groupby('hometown')
        for gp1 in record_gp:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        edgelist.append((nlist[i],nlist[j],{'weight':1}))
        G = nx.MultiGraph()
        G.add_nodes_from(nodelist)
        G.add_edges_from(edgelist)
        nx.set_node_attributes(G, node_attrs)
        nx.write_yaml(G,deten+'/second-layer network/'+deten+gen+'_record_hometown.yaml')
        
        # cause+hometown
        edgelist = []
        record_gp = df.groupby('cause')
        for gp1 in record_gp:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        edgelist.append((nlist[i],nlist[j],{'weight':2}))
        record_gp = df.groupby('hometown')
        for gp1 in record_gp:
            subdf1 = gp1[1]
            nlist = list(subdf1.id)
            if len(nlist) > 1:
                for i in range(len(nlist)-1):
                    for j in range(i+1,len(nlist)):
                        edgelist.append((nlist[i],nlist[j],{'weight':1}))

        G = nx.MultiGraph()
        G.add_nodes_from(nodelist)
        G.add_edges_from(edgelist)
        nx.set_node_attributes(G, node_attrs)
        nx.write_yaml(G,deten+'/second-layer network/'+deten+gen+'_cause_hometown.yaml')
```
